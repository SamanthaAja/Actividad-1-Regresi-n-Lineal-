```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Listings_sin_nulos.csv') 
print(data.info())
```

```python
# Variables cuantitativas int y float
cuantitativas = data.select_dtypes(include=["int64", "float64"])

# Variables cualitativas object
cualitativas = data.select_dtypes(include=["object"])

# Si tambi√©n quieres incluir fechas como cualitativas:
fechas = ["last_scraped", "host_since", "calendar_last_scraped", "first_review", "last_review"]
cualitativas = pd.concat([cualitativas, data[fechas]], axis=1)
```

Procedimiento de cuantiles

```python
#Dividir cuartiles
y=cuantitativas

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Obtenemos datos limpios del Dataframe
data3_iqr= cuantitativas[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data3_iqr
```

```python
valores_nulos=data3_iqr.isnull().sum()
valores_nulos
```

```python
data4_iqr=data3_iqr.copy()
data4_iqr=data4_iqr.fillna(round(data3_iqr.median(),1))
data4_iqr
```

```python
Datos_limpios = pd.concat([cualitativas, data4_iqr], axis=1)
Datos_limpios
```

```python
valores_nulos=Datos_limpios.isnull().sum()
valores_nulos
```

```python
Datos_limpios.to_csv("Listings_sin_atipicos.csv")
```
